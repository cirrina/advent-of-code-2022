```python
## --- Day 5a : Supply Stacks ---
## pop and rock
import re
import csv

write_cargo=True
cargo_dict = {x:[] for x in range(1,10)}

## identify coorfinates (character positions) for the cargo letters
cargo_indexes=[' 1   2   3   4   5   6   7   8   9 '.index(str(i)) for i in range(1,10)]
move_started=False

with open('input-day-5.txt', 'r') as allines:
    # Iterate through the lines in the file
    for row in allines:
        
        # Use a regular expression to find all instances of the pattern "[...]" in the line
        if re.findall(r'\[.*?\]', row):
            cargo=[row[c] for c in cargo_indexes if c < len(row)]
            # [cargo_dict[i].append(c) for i, c in enumerate(cargo, start=1) if not c==' ']
            [cargo_dict[i].append(c) for i, c in enumerate(cargo, start=1) if not c==' ']
        elif re.findall(r'move.*from.*to', row):
            if not move_started: ## when first start reading instructuons, reverse the lists items in crates - this to work seemless as stacks pop/append
               [cargo_dict[i].reverse() for i, c in enumerate(cargo_dict, start=1)]
            move_started=True
            # print(cargo_dict)
            n_boxes = int(re.findall(r'move (\d+) ', row).pop(0))
            from_stack = int(re.findall(r'from (\d+) ', row).pop(0))
            to_stack = int(re.findall(r'to (\d+)', row).pop(0))
            # for each box instance to move,             
            [ cargo_dict[to_stack].append(cargo_dict[from_stack].pop()) for c in  range(n_boxes) ]  
            
            
        else:
            continue

print( ''.join([cargo_dict[c][-1] for c in cargo_dict ] ))
```

```python

## --- Day 5b : Supply Stacks ---
## indexes
import re
import csv

write_cargo=True
cargo_dict = {x:[] for x in range(1,10)}

## identify coorfinates (character positions) for the cargo letters
cargo_indexes=[' 1   2   3   4   5   6   7   8   9 '.index(str(i)) for i in range(1,10)]
move_started=False

with open('input-day-5.txt', 'r') as allines:
    # Iterate through the lines in the file
    for row in allines:
        
        # Use a regular expression to find all instances of the pattern "[...]" in the line
        if re.findall(r'\[.*?\]', row):
            cargo=[row[c] for c in cargo_indexes if c < len(row)]
            # [cargo_dict[i].append(c) for i, c in enumerate(cargo, start=1) if not c==' ']
            [cargo_dict[i].append(c) for i, c in enumerate(cargo, start=1) if not c==' ']
        ## after reading the crates section, expext no more crates. only instructions are allowed to be read
        elif re.findall(r'move.*from.*to', row): # require pattern
            if not move_started: ## when first start reading instructuons, reverse the lists items in crates - this to work seemless as stacks pop/append
               [cargo_dict[i].reverse() for i, c in enumerate(cargo_dict, start=1)]
            move_started=True
            # print(cargo_dict)
            n_boxes = int(re.findall(r'move (\d+) ', row).pop(0))
            from_stack = int(re.findall(r'from (\d+) ', row).pop(0))
            to_stack = int(re.findall(r'to (\d+)', row).pop(0))
            
            # [ cargo_dict[to_stack].append(cargo_dict[from_stack].pop()) for c in  range(n_boxes) ]  
            cargo_dict[to_stack].extend( cargo_dict[from_stack][-n_boxes::] )
            cargo_dict[from_stack]=cargo_dict[from_stack][:-n_boxes]
            
        else:
            continue
# print(cargo_dict)
print( ''.join([cargo_dict[c][-1] for c in cargo_dict ] ))
```
